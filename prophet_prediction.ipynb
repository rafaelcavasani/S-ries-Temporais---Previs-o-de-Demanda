# Séries Temporais - Previsão de Demanda

Iremos ver nesse notebook como fazemos a previsão de demanda da venda de produtos. Com base em dados históricos, iremos prever a quantidade de produtos que serão vendidos nos próximos três meses.

## Base de Dados

Utilizaremos um dataset do kaggle: "Store Item Demand Forecasting Challenge". Esse dataset contém dados históricos de vendas de 50 produtos em 10 lojas diferentes. Os dados são de vendas realizadas entre janeiro de 2013 e dezembro de 2016.

Link:
<https://www.kaggle.com/c/demand-forecasting-kernels-only>

### Atributos

* **date**: Data da venda;
* **store**: Identificação da loja que foi realizada a venda;
* **item**: Produto que foi vendido;
* **sales**: Quantidade de itens vendidos na data. Esse é o atributo que queremos prever;

## Importando as Bibliotecas

Utilizaremos as bibliotecas **pandas** para importar os dados e **fbprophet** para realizar as previsões de vendas

```python
import pandas as pd
from fbprophet import Prophet
```

## Importando o Banco de Dados

A função **read_csv**, do pandas, lê o arquivo csv e transforma em um objeto **DataFrame** para trabalharmos com os dados.

```python
data = pd.read_csv('train.csv')
```

## Treinamento e Previsões

Queremos treinar o modelo e predizer a quantidade das vendas de cada produto em cada loja separadamente.

Para isso, iremos fazer dois laços de repetição para conseguir iteragir com todos os 50 itens e todas as 10 lojas do dataset. Com isso, em cada iteração do segundo laço, teremos um conjunto com número de produto e loja diferentes. 

Com os números em mãos, iremos filtrar no dataset somente as vendas do produto e loja da iteração, com o comando **query**, passando o resultado para um novo DataFrame, instanciado como data_predict.

Utilizaremos o Prophet para fazer as predições. O Prophet é um pacote do python, criado pelo facebook, utilizado para previsões de séries temporais.

Para utilizá-lo, temos que criar um DataFrame com as colunas:

* **ds**: Data das vendas;
* **y**: Quantidade de vendas ou valores que deseja prever;

Instanciamos o modelo com a classe Prophet e chamamos o método **fit** para treinar o modelo.

Após isso, criamos um DataFrame com a quantidade de períodos que queremos prever e a frequência dos dados. Em nosso caso, 90 dias com frequência  diária. E chamamos o método **predict**, passando o DataFrame criado como parâmetro.

Faremos outro laço de repetição para adicionar as predições a lista **sales_list**, conseguindo, assim, juntar todas as predições de todos os produtos.

Por fim, transformamos todas as nossas predições em um DataFrame, para exportarmos para um arquivo csv e enviarmos ao kaggle.

```python
id_list = []
sales_list = []
id = 0
for item in range(1, 51):
    for store in range(1, 11):
        print('store: {} - item: {}'.format(store, item))
        data_predict = data.query("item == @item & store == @store")
        X = pd.DataFrame(index=range(0, len(data_predict)))
        X['ds'] = data_predict['date'].values
        X['y'] = data_predict['sales'].values
        m = Prophet()
        m.fit(X)
        future = m.make_future_dataframe(periods=90, freq='D')
        forecast = m.predict(future)
        
        for pred in forecast['yhat']:
            id_list.append(id)
            sales_list.append(pred)
            id += 1
            
prediction_df = pd.DataFrame({'id': id_list, 'sales': sales_list})
```

```python
prediction_df
```

## Preparando para Enviar ao Kaggle

```python
final_df = prediction_df
final_df['sales'] = final_df['sales'].astype('int32')
final_df.to_csv('submission_int.csv', index=False)
```

Para enviar ao kaggle, precisamos transformar os valores "quebrados" em valores inteiros, sem casas decimais. Para isso, iremos iterar sobre todas as linhas do DataFrame e, se o valor após a casa decimal ultrapassar 0.5, iremos somar 1 ao valor, arredondando o valor para cima. Caso contrário, arredondamos para baixo, simplesmente ignorando as casas decimais. 

```python
final_float = prediction_df
for index, row in final_float.iterrows():
    sales = row['sales']
    rest = sales - row['sales'].astype('int32')
    if rest > 0.5:
        sales += 1
    
    final_float.at[index, 'sales'] = sales
    
final_float.to_csv('submission_float.csv', index=False)
```

Por fim, exportamos o DataFrame para um arquivo csv chamado de submission.csv com o método **to_csv** do Pandas.

```python
final_float['sales'] = final_float['sales'].astype('int32')
final_float.to_csv('submission_float.csv', index=False)
```
